```python
import networkx as nx
import urllib.request
import gzip
import os

url = "https://snap.stanford.edu/data/facebook_combined.txt.gz"
filename = "facebook_combined.txt"

# Télécharger et décompresser si le fichier n'existe pas
if not os.path.exists(filename):
    print("Téléchargement du dataset Facebook...")
    urllib.request.urlretrieve(url, "facebook_combined.txt.gz")
    with gzip.open("facebook_combined.txt.gz", 'rb') as f_in:
        with open(filename, 'wb') as f_out:
            f_out.write(f_in.read())
    os.remove("facebook_combined.txt.gz")
    print("Téléchargement terminé!")

# Charger le graphe
G = nx.read_edgelist(filename, nodetype=int)
print(f"Graphe chargé: {G.number_of_nodes()} nœuds, {G.number_of_edges()} arêtes")
```

```python
import matplotlib.pyplot as plt
import random
import networkx as nx

# Créer un sous-échantillon pour visualisation (300 nœuds max)
start_node = random.choice(list(G.nodes()))
bfs_nodes = list(nx.bfs_tree(G, start_node))[:300]
G_sample = G.subgraph(bfs_nodes)

# Visualisation
plt.figure(figsize=(14, 10))
pos = nx.spring_layout(G_sample, seed=42, k=2)

# Dessiner les arêtes
nx.draw_networkx_edges(G_sample, pos, alpha=0.3, edge_color='gray', width=0.8)

# Dessiner les nœuds (taille selon le degré)
node_sizes = [G_sample.degree(node) * 8 for node in G_sample.nodes()]
nx.draw_networkx_nodes(G_sample, pos, node_size=node_sizes, 
                      node_color='skyblue', alpha=0.7, 
                      edgecolors='black', linewidths=0.5)

plt.title(f"Échantillon du réseau Facebook ({G_sample.number_of_nodes()} nœuds)")
plt.axis('off')
plt.tight_layout()
plt.show()

print(f"Visualisation de {G_sample.number_of_nodes()} nœuds et {G_sample.number_of_edges()} arêtes")
```

```python
from networkx.algorithms.community import girvan_newman
import matplotlib.pyplot as plt
import networkx as nx

# Étape 1: Exécuter Girvan-Newman
communities = girvan_newman(G)
first_level_communities = list(next(communities))  # Premier niveau de division

# Étape 2: Créer un dictionnaire de couleurs pour chaque nœud
community_colors = {}  # ← On crée la variable manquante !
colors = ['red', 'blue', 'green', 'orange', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan']

# Assigner une couleur à chaque communauté
for i, community in enumerate(first_level_communities):
    for node in community:
        community_colors[node] = colors[i % len(colors)]

# Étape 3: Visualiser
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G)

# Maintenant community_colors existe !
nx.draw(G, pos, 
        node_color=[community_colors[node] for node in G.nodes()], 
        with_labels=False, 
        edge_color='lightgray',
        node_size=50)

plt.title("Community Visualization (Girvan-Newman)")
plt.axis('off')
plt.show()

# Afficher les communautés
print("Communities (Girvan-Newman):")
for i, community in enumerate(first_level_communities):
    print(f"Communauté {i+1} ({len(community)} nœuds): {list(community)[:10]}...")
```

```python
density = nx.density(G)
print("Graph Density:", density)
if nx.is_connected(G):
    diameter = nx.diameter(G)
    radius = nx.radius(G)
    print("\nNetwork Diameter:", diameter)
    print("\nNetwork Radius:", radius)
else:
    print("\nGraph is not connected; diameter and radius are undefined.")
global_clustering = nx.average_clustering(G)
print("\nGlobal Clustering Coefficient:", global_clustering)
k_core = nx.k_core(G)
print("\nk-Core Decomposition:", k_core.nodes())
local_clustering = nx.clustering(G)
print("\nLocal Clustering Coefficient:", local_clustering)
```
